```python
import pandas as pd
import numpy as np
```

### Parameters  of PivotTables

* Data
* Values
* Index
* Columns
* Aggfunc
* Fill_values
* Margins

```python
x = { 'Company': ['Ford', 'Ford', 'VW', 'Audi', 'Audi', 'Ford'],
     'Price': [2, 3, 1, 2, 3, 1],
     'Weight': [4, 6, 2, 2, 4, 3],
     'Length': [6, 7, 3, 6, 8, np.nan], # np <- from Numpy Library; nan <- Not a Numeric Value
     'Made': ['China', 'Mexico', 'Mexico', 'Corea', 'UK', 'UK'],
    }
df = pd.DataFrame(data = x)
df
```

### Check if the data is good

```python
df.isnull().values.any()
```

```python
df.isnull().sum()
```

### Check each column, check which one has the problem

```python
df['Price'].isnull().values.any()
```

```python
df['Made'].isnull().values.any()
```

```python
df['Weight'].isnull().values.any()
```

```python
df['Length'].isnull().values.any()
```

```python
df['Length'].fillna(value = df['Length'].mean(),
                    inplace = True
                   )
df
```

### What is the price of audi cars made in china?

```python
pd.pivot_table(df,
               index = 'Company', 
               values = ['Price', 'Weight'],
               aggfunc = {'Price':np.mean, 
                          'Weight':np.sum},
               columns = 'Made',
               margins = True, 
               margins_name = '||**GrandTotal**||'
              )
```
